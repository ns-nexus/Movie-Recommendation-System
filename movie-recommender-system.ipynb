## Data Preprocessing

```python
# merging movies and credits dataset to create new dataset
data = credits.merge(movies,on='title')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
# selecting informative columns from the dataset
data = data[['movie_id','title','overview','popularity','genres','keywords','cast','crew']]
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# removing null values
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
# checking for duplicacy
data.duplicated().sum()
```

```python
data['overview'] = data['overview'].apply(lambda x:x.split())
data.head()
```

```python
# converting dictionary into list
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
data['genres'] =data['genres'].apply(convert)
data['keywords'] =data['keywords'].apply(convert)
data.head()
```

```python
# fetching name of top 3 cast member names and converting into list
def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
data['cast'] = data['cast'].apply(convert_cast)
data.head()
```

```python
# fetching director name from crew column and converting into list
def convert_crew(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
data['crew'] = data['crew'].apply(convert_crew)
data.head()
```

```python
# removing space between words
data['genres'] = data['genres'].apply(lambda x: [i.replace(" ","") for i in x])
data['keywords'] = data['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
data['cast'] = data['cast'].apply(lambda x: [i.replace(" ","") for i in x])
data['crew'] = data['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
data.head()
```

```python
# concatinate overview, genres, keywords, cast and crew column to create a new column 'tags'
data['tags'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew']
new_data = data.drop(columns=['overview','genres','keywords','cast','crew'])
new_data.head()

# convert tags column data into string
new_data['tags'] = new_data['tags'].apply(lambda x: " ".join(x))

# convert tags into lowercase
new_data['tags'] = new_data['tags'].apply(lambda x: x.lower())

# final dataset
new_data.head()
```

## Text Vectorization

```python
# apply stemming (reduction to root word)
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
new_data['tags'] = new_data['tags'].apply(stem)
```

```python
# vectorizing 5000 most frequent words and removing stop words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_data['tags']).toarray()

# vectors in the first movie
vectors[0]
```

```python
# 5000 keywords
feature_names = cv.get_feature_names_out()
np.array2string(feature_names, threshold=np.inf)
```

## Model Creation

```python
# finding similarity between vectors using cosine similarity function
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
# function to recommend 10 similar movies
def recommend(movie):
    index = new_data[new_data['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
